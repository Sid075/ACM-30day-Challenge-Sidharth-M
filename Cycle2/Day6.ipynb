```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

from sklearn.ensemble import RandomForestClassifier, AdaBoostClassifier
from xgboost import XGBClassifier
from sklearn.metrics import confusion_matrix
```

```python
file_path = "/kaggle/input/breast-cancer-dataset/breast-cancer.csv"
df = pd.read_csv(file_path)
```

```python
df.head()
```

```python
print(df.columns)
```

```python
print(df['diagnosis'].unique())
```

```python
df['diagnosis'] = df['diagnosis'].map({'M': 1, 'B': 0})

# null values 
df.isnull().sum()
df = df.dropna()
```

```python
X = df.drop('diagnosis', axis=1)
y = df['diagnosis']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=69)
```

```python
# Random Forest
rf = RandomForestClassifier(random_state=42)
rf.fit(X_train, y_train)
rf_pred = rf.predict(X_test)

# AdaBoost
ab = AdaBoostClassifier(random_state=42)
ab.fit(X_train, y_train)
ab_pred = ab.predict(X_test)

# XGBoost
xgb = XGBClassifier(use_label_encoder=False, eval_metric='logloss', random_state=42)
xgb.fit(X_train, y_train)
xgb_pred = xgb.predict(X_test)
```

```python
print("Accuracy Comparison")
print("Random Forest Accuracy     :", accuracy_score(y_test, rf_pred))
print("Random Forest Confusion Matrix:\n", confusion_matrix(y_test, rf_pred), "\n")

print("AdaBoost Accuracy           :", accuracy_score(y_test, ab_pred))
print("AdaBoost Confusion Matrix:\n", confusion_matrix(y_test, ab_pred), "\n")

print("XGBoost Accuracy            :", accuracy_score(y_test, xgb_pred))
print("XGBoost Confusion Matrix:\n", confusion_matrix(y_test, xgb_pred), "\n")
```

BEST MODEL IS ADABOOST
